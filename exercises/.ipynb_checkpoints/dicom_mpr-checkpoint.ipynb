# 3D Volume MPR

```python
import pydicom
import numpy as np
from PIL import Image
import os
import matplotlib.pyplot as plt
```

### Read dcm slices

```python
# directory of stored dcm files
dcm_path = 'volume'

# read all slices
slices = [pydicom.dcmread(os.path.join(dcm_path, f)) for f in os.listdir(dcm_path)]

# sort the slices by the ImagePositionPatient[0]
slices = sorted(slices, key=lambda x: x.ImagePositionPatient[0])
```

```python
# number of slices
print(len(slices))
```

### Pixel data

```python
pixel_data = np.stack([s.pixel_array for s in slices])
print(pixel_data.shape)
```

### Adjust the aspect ratio

```python
print("Slice thickness = {}".format(slices[0].SliceThickness))
print("Pixel spacing = {}".format(slices[0].PixelSpacing))
print("Aspect ratio = {:.4f}".format(slices[0].SliceThickness / slices[0].PixelSpacing[0]))
```
